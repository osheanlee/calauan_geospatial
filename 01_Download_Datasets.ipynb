```python
import pandas as pd
import geopandas as gpd
import requests
import zipfile
import os
```

# 01 - Create requests function to download dataset

```python
def download_dataset(url, download_path, extraction_dir, filename):

  # Create the extraction directory if it doesn't exist
  os.makedirs(extraction_dir, exist_ok=True)

  # Download the zip file
  print(f"Downloading {url}...")
  response = requests.get(url, stream=True)
  response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)

  with open(download_path, "wb") as f:
      for chunk in response.iter_content(chunk_size=8192):
          f.write(chunk)
  print("Download complete.")

  # Unzip the file if it is a zipfile, otherwise skip
  if download_path.endswith('.zip'):
    print(f"Extracting {download_path} to {extraction_dir}...")
    with zipfile.ZipFile(download_path, 'r') as zip_ref:
        zip_ref.extractall(extraction_dir)
    print("Extraction complete.")

    # Clean up the downloaded zip file
    os.remove(download_path)
    print(f"Removed downloaded zip file: {download_path}")

  else:
    print(f"Skipping extraction for non-zip file: {download_path}")
```

# 02a - Download Calauan ADM4 (Barangay Boundaries)

```python
# Define the URL for the zip file and the target directory
url = "https://data.humdata.org/dataset/caf116df-f984-4deb-85ca-41b349d3f313/resource/4e4067d6-3fc8-4d2c-91ea-330c3d4f8e8e/download/phl_admin_boundaries.shp.zip"
download_path = "./phl_admin_boundaries.shp.zip"
extraction_dir = "./phl_admin_boundaries_data"
filename = "phl_admin_boundaries.zip"

download_dataset(url, download_path, extraction_dir, filename)
```
