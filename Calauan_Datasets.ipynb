```python
import geopandas as gpd
import os
import pandas as pd
import requests
import zipfile
```

# Admin Boundaries of Calauan, Laguna
Get from HDX

```python
zipfile_url = "https://data.humdata.org/dataset/caf116df-f984-4deb-85ca-41b349d3f313/resource/12457689-6a86-4474-8032-5ca9464d38a8/download/phl_adm_psa_namria_20231106_shp.zip"

# Define the download and extraction paths
download_path = "phl_adm_psa_namria_20231106_shp.zip"
extract_path = "extracted_data"

# Download the zip file
response = requests.get(zipfile_url)
with open(download_path, "wb") as zip_file:
    zip_file.write(response.content)

# Extract the contents of the zip file
with zipfile.ZipFile(download_path, "r") as zip_ref:
    zip_ref.extractall(extract_path)

# Optionally, remove the downloaded zip file
os.remove(download_path)

print(f"Downloaded and extracted to {extract_path}.")
```

## Extract Calauan and Barangay Boundaries

```python
phl_gdf = gpd.read_file('/content/extracted_data/phl_admbnda_adm4_psa_namria_20231106.shp')
```
